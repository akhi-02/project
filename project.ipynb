### Number of orders for different months

```python
df_new['monthly_order']=df_new['invoice_date'].dt.month
df_new.groupby('monthly_order')['invoice_num'].nunique()
```

### Plot the graph for the above query

```python
import matplotlib.pyplot as plt
month=['jan','feb','mar','apr','may','june','july','aug','sept','oct','nov','dec']
order=[1236,1202,1619,1384,1849,1707,1593,1544,2078,2263,3086,2629]
plt.plot(month,order,'k',linewidth=3,marker='o',markersize=8)
plt.xlabel('month')
plt.ylabel('order')
plt.title('monthly_order')
plt.show()
          
```

### How many orders for each country?

### Draw a horizontal bar graph for above
